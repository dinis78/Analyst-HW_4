*Особо год постройки не влияет на цену домов*

```python
corr_matrix = df.corr()
corr_matrix = np.round(corr_matrix, 1)
corr_matrix[np.abs(corr_matrix)<0.3] = 0
corr_matrix
```

```python
sns.heatmap(corr_matrix, annot=True, linewidths=5, cmap='coolwarm')
plt.title('Тепловая карта');
```

*Вывод: в основном цена домов зависит от жилой площади, количесва комнат, вида и класса дома.*
